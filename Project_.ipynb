```python
import pandas as pd
```

Step-1 : Importing the data from csv file, by giving proper path and rewriting "\" into "/".

```python
df=pd.read_csv("C:/Users/tejas/PycharmProjects/JupyterProject/data/Dataset .csv")
```

Dataset "df" before cleaning.


```python
df
```

Step-2 : Rewriting and cleaning the columns names for better SQL querying.

```python
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
```

Step-3 : Remove duplicate rows if present.

```python
df.drop_duplicates(inplace=True)
```

Step-4 : Object to String conversion and white spaces trimming.

```python
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].astype(str).str.strip()
```

Step-5 : Converting numeric columns properly.

```python
num_cols = ["votes", "aggregate_rating", "longitude", "latitude", "price_range", "average_cost_for_two"]
for col in num_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
```

Step-6 : Replace invalid values with NaN


```python
df.replace(["-", "--", "NaN", "nan"], pd.NA, inplace=True)
```

Step-7 : Saving cleaned data and deleting index values from data.

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
